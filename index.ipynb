## Question 2

Explain the relationship between **Primary Keys** and **Foreign Keys**.

Type your answer below this line:
_______________________________________________________________________________________________________________________________

Primary keys uniquely identify rows within a table. Foreign keys identify rows for tables existing outside of the current table. A foreign key from one table can match up with the primary key within another table. This is necessary to creating JOIN statements within SQL framework. 



## Question 3

Explain the different types of relationships entities can have in a SQL database. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________

One-to-One: records line up exactly 1 to 1 across 2 or more tables.
One-to-Many: records within 1 table match up with multple records within  another table i.e. 1 fish in a table is a compenent of another table, the school, which houses multple fish.
Many-to-Many: each data point within a dataset match up with multple data points of another dataset and vica-versa.

## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________

INNER JOIN: Returns data where keys match/overlap
OUTER JOIN:Returns data where keys do not match/overlap
LEFT JOIN: Returns data from left dataset and any matching keys within the right data set.
RIGHT JOIN: Returns data from right dataset and any matching keys within the left data set.
